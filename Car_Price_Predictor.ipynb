MODEL

```python
x=cars_train.drop(columns='Price')
y=cars_train['Price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
ohe=OneHotEncoder()
ohe.fit(x[['Name','Company','Fuel']])
```

```python
column=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['Name','Company','Fuel']),remainder='passthrough')
```

```python
model_linear=LinearRegression()
```

```python
pipe=make_pipeline(column,model_linear)
```

```python
pipe.fit(x_train,y_train)
```

```python

y_predict=pipe.predict(x_test)
```

```python
y_predict
```

```python
r2_score(y_test,y_predict)
```

```python
scores=[]
for i in range(1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    model_linear=LinearRegression()
    pipe=make_pipeline(column,model_linear)
    pipe.fit(x_train,y_train)
    y_predict=pipe.predict(x_test)
    scores.append(r2_score(y_test,y_predict))
```

```python
scores[np.argmax(scores)]
```

```python
np.argmax(scores)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=np.argmax(scores))
model_linear=LinearRegression()
pipe=make_pipeline(column,model_linear)
pipe.fit(x_train,y_train)
y_predict=pipe.predict(x_test)
r2_score(y_test,y_predict)
```
